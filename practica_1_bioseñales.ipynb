```python
#Repaso Pandas

# 1) Recibe un diccionario con las notas de los alumnos y devuelve una serie
#    con la nota mnima, mxima, media y desviacin estndar.
import pandas as pd
import numpy as np

#la funcion recibe un diccionario
def estadisticas_notas(notas):

    valores = list(notas.values())  # Extraer las nota recorriendo el diccionario

    # Crear una serie con las estadsticas
    resultados = pd.Series({
        "Nota mnima": np.min(valores), #Valor minimo
        "Nota mxima": np.max(valores), #valor maximo
        "Media": np.mean(valores), #promedio
        "Desviacin estndar": np.std(valores, ddof=1)  # desviacion estandar ddof=1 para la muestra
    })

    return resultados

# Diccionario con las notas de los alumnos
notas_alumnos = {
    "Manu": 4.2,
    "Vale": 3.6,
    "Isa": 4.6,
    "Andre": 3.8,
    "Omar": 3.9,
    "Juli": 2.9,
    "Felipe": 5.0
}

resultado = estadisticas_notas(notas_alumnos)
print(resultado.to_string()) # .to_spring() para que no salga el tipo de variable
```

```python
#Punto 2 Repaso Panda
import pandas as pd
import os
from google.colab import files
uploaded = files.upload()
# Punto a Cargar los datos
file_path = "datos.csv"  # Google Colab lo guarda en la raz
datos = pd.read_csv(file_path, sep=";")

# Punto b Verificar que se cargaron los datos
print(" Archivo cargado correctamente")
print(datos.head())  # Muestra las primeras filas
print(datos.tail()) #Muestra las ultimas

#Punto c Eliminar la columna unnamed
if "Unnamed: 0" in datos.columns:
    datos.drop(columns=["Unnamed: 0"], inplace=True)

#convertir de centimetros a metros
datos["Weight"]= datos["Weight"]/100

# Punto D  Calcular el IMC
datos["BMI"] = datos["Height"] / (datos["Weight"] ** 2)

#Punto F: categoizar
def categorizar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc < 25:
        return "Normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

# Punto E renombrar como BMI
datos["BMI_Category"] = datos["BMI"].apply(categorizar_imc)

# Mostrar el resultado
print(datos.head(40))
```
